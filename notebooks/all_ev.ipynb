```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
# This line makes matplotlib figures render in the notebook
%matplotlib inline
```

```python
# Define colors and resolution for figures
dip_color = '#FFE800'
tet_color = '#A600A6'
comb_color = '#1B1BB3'

reg_color = '#ff4293'
DPI = 600
```

```python
# Read in the data
go = pd.read_csv('../data/final_figures_input/GO_EV.txt', sep='\t')
mn = pd.read_csv('../data/final_figures_input/AraCyc_EV.txt', sep='\t')
tas = pd.read_csv('../data/final_figures_input/Tasdigh_EV.txt', sep='\t')


# Color palette stuff
class_colors = [dip_color, tet_color, comb_color]
sns.set_palette(class_colors)
sns.palplot(sns.color_palette())
```

```python
go = go.melt("Class", var_name="Ploidy", value_name="EV")
mn = mn.melt("Class", var_name="Ploidy", value_name="EV")
tas = tas.melt("Rank", var_name="Ploidy", value_name="EV")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(8,8), dpi=DPI, sharex=True)

#################################
####       GO EV Plots       ####
#################################

ax1 = sns.boxenplot(data=go,
                    y="Class",
                    x="EV",
                    hue="Ploidy",
                    orient="h",
                    palette=class_colors,
                    ax=ax1)

ax1.get_xaxis().set_tick_params(direction='out')
ax1.xaxis.set_ticks_position('bottom')
ax1.set_xlim(0,1.15)
ax1.set_ylabel(' ')
ax1.set_xlabel(' ')

labels = ['Class I', 'Class II']
ax1.set_yticklabels(labels, rotation=45)

# label_diff(ax1, 0, 1,'$p=x$',ind,menMeans, y_shift=-0.4, y_diff=.44)
# label_diff(ax1, 2, 3,'$p=y$',ind,menMeans, y_shift=-0.4, y_diff=.44)
# label_diff(ax1, 4, 5,'$p=z$',ind,menMeans, y_shift=-0.4, y_diff=.44)

#################################
####       MN EV Plots       ####
#################################
ax2 = sns.boxenplot(data=mn,
                    y="Class",
                    x="EV",
                    hue="Ploidy",
                    orient="h",
                    palette=class_colors,
                    ax=ax2)

ax2.get_xaxis().set_tick_params(direction='out')
ax2.xaxis.set_ticks_position('bottom')
ax2.set_xlim(0,1.15)
ax2.set_ylabel(' ')
ax2.set_xlabel(' ')

ax2.set_yticklabels(labels, rotation=45)

##################################
####       Tas EV Plots       ####
##################################
ax3 = sns.boxenplot(data=tas,
                    y="Rank",
                    x="EV",
                    hue="Ploidy",
                    orient="h",
                    palette=class_colors,
                    ax=ax3)

ax3.get_xaxis().set_tick_params(direction='out')
ax3.xaxis.set_ticks_position('bottom')
ax3.set_xlim(0,1.15)
ax3.set_ylabel(' ')
ax3.set_xlabel('EV')
ax3.legend(title="Ploidy", loc="upper right")

labels = ["Bottom 1,000", "Top 1,000"]
ax3.set_yticklabels(labels, rotation=45)

# General plot tidying
plt.subplots_adjust(bottom=0.0, wspace=0.15)
plt.savefig('../figures/EV_boxen.png')
```
